```python
hospital_data = pd.read_csv('heart-disease.csv')
hospital_data.head(10)
```

```python
hospital_data.age.plot.hist(bins=100);
```

```python
hospital_data.plot.hist(subplots=True, figsize=(10,100), bins=20);
```

```python
hospital_data
```

```python
over_50 = hospital_data[hospital_data['age']>50]
over_50
```

```python
over_50.plot(kind='scatter', y='chol', x='age', c='y')
```

```python
fig, ax = plt.subplots(figsize=(4,4))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45, 90])
```

```python
fig, ax = plt.subplots(figsize=(4,4))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='m',
            ax=ax);
ax.set_xlim([45, 90])
```

```python
hospital_data.head()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, figsize=(10,10))
scatter1 = ax1.scatter(x=hospital_data.age, y=hospital_data.chol, c=hospital_data.target, cmap='summer')
ax1.set(title='Age/Cholesterol',
       ylabel='Cholesterol',
       xlabel='Age'
      )
ax1.legend(*scatter1.legend_elements(), title='Targets', loc='lower right')
ax1.axhline(hospital_data.chol.mean(), c='m', ls=':');

scatter2 = ax2.scatter(x=hospital_data.age, y=hospital_data.thalach, c=hospital_data.target, cmap='winter')
ax2.set(title='Age/Heart Rate',
       xlabel='Age',
       ylabel='Hearth rate',
        xlim=(20,80)
       )
ax2.legend(*scatter2.legend_elements(), title='Targets', loc='lower left')
ax2.axhline(hospital_data.thalach.mean(), c='m', ls=':');

fig.suptitle('UBTH Data Map', fontsize=20, fontweight='bold')
```
